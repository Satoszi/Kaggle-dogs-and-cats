```python
import numpy as np
import shutil
import os
import matplotlib.pyplot as plt
import cv2
```

```python
def move_abnormal_examples(dataset_directory, bad_examples):
    bad_cats = [img_name for img_name in bad_examples if "cat" in img_name]
    bad_dogs = [img_name for img_name in bad_examples if "dog" in img_name]
    
    for bad_cat in bad_cats:
        try:
            shutil.move(os.path.join(dataset_directory,"training_set", "cats", bad_cat),
                    os.path.join(dataset_directory,"abnormal", "cats", bad_cat))
        except: None
    for bad_dog in bad_dogs:
        try:
            shutil.move(os.path.join(dataset_directory,"training_set", "dogs", bad_dog),
                    os.path.join(dataset_directory,"abnormal", "dogs", bad_dog))
        except: None
```

```python
def show_abnormal(label):

    bad_images = []
    for i in os.listdir(os.path.join(dataset_directory,"abnormal",label)):
        image = cv2.imread(os.path.join(dataset_directory,"abnormal",label, i))
        bad_images.append(image)

    fig = plt.figure(figsize=(10, 7))

    for i in range(9):
        fig.add_subplot(3, 3, i+1)
        plt.imshow(bad_images[i])
```

```python
bad_examples = []
with open('abnormal.txt') as f:
    bad_examples = f.read().splitlines()

bad_examples = np.unique(bad_examples)
```

```python
dataset_directory = "C:\\Users\\Tazz\\Desktop\\AGH\\Moje\\ML\\Kaggle-dogs-and-cats\\dataset"
move_abnormal_examples(dataset_directory, bad_examples)
```
